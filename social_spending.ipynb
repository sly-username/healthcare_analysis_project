```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
health_social = pd.read_csv('data/Bradley/health_social.csv')
```

```python
health_social.head(3)
```

```python
cumval=0
fig = plt.figure(figsize=(30, 10))
for col in health_social.columns[~health_social.columns.isin(['Country'])]:
    plt.bar(health_social['Country'], health_social[col], bottom=cumval, label=col)
    cumval = cumval+health_social[col]
```
