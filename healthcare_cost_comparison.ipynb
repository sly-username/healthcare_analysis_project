```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
prices = pd.read_csv('data/IFHP/prices.csv')
prices = prices.rename(index=str, columns={'type': 'procedure'})
```

```python
prices['intervention'].unique()
```

```python
plt.figure()
procedure_prices = prices.groupby('intervention')

for key, grp in procedure_prices:
    title = ''.join((grp.procedure[0], ': ' + key))
    grp.plot(kind='barh', x='country', y='value', title=title, 
             color=np.where(grp['country'] == 'US', 'r', 'grey'), 
             legend=False)
```
